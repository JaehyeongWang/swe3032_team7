```python
import yfinance as yf
from alpha_vantage.fundamentaldata import FundamentalData
import json

def get_btm_ratio(symbol, av_api_key):
    # Fetch market data using yfinance
    ticker = yf.Ticker(symbol)
    market_cap = ticker.info['marketCap']
    
    # Fetch book value using Alpha Vantage
    fd = FundamentalData(av_api_key, output_format='json')
    bs_data, _ = fd.get_balance_sheet_annual(symbol)
    
    # Get the most recent balance sheet data
    total_assets = float(bs_data.totalAssets[0])
    total_liabilities = float(bs_data.totalLiabilities[0])
    
    # Calculate book value of equity
    book_value = total_assets - total_liabilities
    
    # Calculate book to market ratio
    btm_ratio =  book_value / market_cap
    
    return btm_ratio

av_api_key = 'Q510M2GPL619BYUD'
```

```python
import random
```

```python
def bm_prepro(type):
    with open(f'{type}_bm.txt', 'r') as f:
        lines = f.readlines()

    prepro_list = [line for idx, line in enumerate(lines) if idx % 2 == 1]
    prepro_list = [float(line.strip()) for line in prepro_list]
    prepro_list = [value for value in prepro_list if value >= 0]
    prepro_list = random.sample(prepro_list, 400)
    return prepro_list
```

```python
growth_list = bm_prepro('growth')
overlap_list = bm_prepro('overlap')
value_list = bm_prepro('value')
```

```python
labeled_values = [(v, 'growth') for v in growth_list] + [(v, 'overlap') for v in overlap_list] + [(v, 'value') for v in value_list]

labeled_values.sort()

sorted_values = [v for v, _ in labeled_values]

sorted_growth = sorted_values[:400]
sorted_overlap = sorted_values[400:800]
sorted_value = sorted_values[800:]

sorted_labels = [label for _, label in labeled_values]
growth_labels = sorted_labels[:400]
overlap_labels = sorted_labels[400:800]
value_labels = sorted_labels[800:]
```

```python
correct_growth = sum([1 for v, l in zip(sorted_growth, growth_labels) if l == 'growth'])
correct_overlap = sum([1 for v, l in zip(sorted_overlap, overlap_labels) if l == 'overlap'])
correct_value = sum([1 for v, l in zip(sorted_value, value_labels) if l == 'value'])

total_correct = correct_growth + correct_overlap + correct_value

accuracy = total_correct / len(labeled_values)
accuracy
```
