```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Replace 'YOUR_INDEX_SYMBOL' with the index symbol you want to collect financial information for (e.g. 'SPX' for S&P 500)
index_symbol = 'IWF'

# Set up the Selenium driver (make sure you have installed the appropriate driver for your browser)
driver = webdriver.Chrome()

# Navigate to the Yahoo Finance page for the index
driver.get(f'https://finance.yahoo.com/quote/{index_symbol}')

# Wait for the page to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//div[contains(@class,"P(10px)")]')))

# Find the list of stocks in the index
stock_list = driver.find_elements(By.XPATH, '//table[contains(@class,"W(100%)")]//tr[contains(@class,"BdT") and not(contains(@class,"C($linkColor)"))]')

# Loop through each stock in the list and collect its financial information
for stock in stock_list:
    # Extract the stock symbol from the list
    stock_symbol = stock.find_element(By.XPATH, './/td[1]/a').text
    
    # Click on the stock symbol to navigate to its financial information page
    stock_link = stock.find_element(By.XPATH, './/td[1]/a').get_attribute('href')
    driver.get(stock_link)
    
    # Wait for the financial information to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//div[contains(@class,"Mstart(0)")]')))
    
    # Extract the financial information from the page (replace these XPATHs with the ones specific to your desired financial data)
    price = driver.find_element(By.XPATH, '//span[contains(@class,"Trsdu(0.3s)")]')
    market_cap = driver.find_element(By.XPATH, '//td[contains(text(),"Market Cap")]/following-sibling::td')
    pe_ratio = driver.find_element(By.XPATH, '//td[contains(text(),"PE Ratio (TTM)")]/following-sibling::td')
    
    # Print the financial information for the stock
    print(f'Stock: {stock_symbol} | Price: {price.text} | Market Cap: {market_cap.text} | P/E Ratio: {pe_ratio.text}')
    
    # Navigate back to the index page for the next stock
    driver.back()

# Close the browser
driver.quit()
```
